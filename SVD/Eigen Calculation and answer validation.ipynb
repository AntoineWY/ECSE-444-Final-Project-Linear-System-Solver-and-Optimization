```python
from numpy import linalg as LA
from numpy import array
from math import sqrt
import numpy as np


a = array([[1.025,2.09142,-0.313,6.1,5.78],[1.24,-2.435,1.3324,2.67,8.114],[7.01,1.2048,-1.245,5.60931,1.335],\
           [3.214,-6.221,0.9243,-1.693,7.2431],[1.4632,4.1231,3.25,-4.2123,1.9938]])
b = array([1,2,5,3,1])
print (a.T.dot(a) )
print ( LA.eig((a.T.dot(a)))  )
print("------------------------below is svd decomposition---------------------")


u,s,vt = LA.svd(a)


print(vt.T)
print(u)



print("\n------------------------below is solving the linear system---------------------")
x = LA.solve(a,b)
print(x)


print("\n------------------------below is solving the linear system by svd components ---------------------")
diag_eigenvector = array([[1/sqrt(216.37561647),0,0,0,0],\
                          [0,1/sqrt(100.08410035),0,0,0],\
                          [0,0,1/sqrt(49.38342505),0,0],\
                          [0,0,0,1/sqrt(34.17658276),0],\
                          [0,0,0,0,1/sqrt(0.35340114)]])
x_svd = vt.T.dot(diag_eigenvector).dot(u.T).dot(b)
print(x_svd)
```
